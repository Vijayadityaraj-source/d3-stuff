```python
distances = np.linalg.norm(df - kmeans.cluster_centers_[kmeans.labels_], axis=1)
# Add distance to DataFrame
df['distance_from_center'] = distances
df['clusters'] = pd.Series(kmeans.labels_)
df.head()
```

```python
sns.countplot(x=df['clusters'])
plt.show()
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=3)
x = pca.fit_transform(df.drop('clusters',axis=1))
```

```python
# Loadings
loadings = pca.components_

dfn=df.drop('clusters',axis=1)
# Get feature names from your DataFrame
feature_names = dfn.columns

# Create a DataFrame for the loadings
loadings_df = pd.DataFrame(loadings, columns=feature_names, index=[f'PC{i+1}' for i in range(loadings.shape[0])])

print("Loadings of Principal Components:")
print(loadings_df)
```

```python
# For each principal component, find the most influential variables
for i in range(loadings.shape[0]):
    print(f"\nPrincipal Component {i+1}:")
    component_loadings = loadings_df.loc[f'PC{i+1}']
    sorted_loadings = component_loadings.abs().sort_values(ascending=False)
    print("Most influential variables:")
    print(sorted_loadings.head(10))  # Display top 10 most influential variables
```

```python
# If needed, you can visualize the loadings
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
sns.heatmap(loadings_df, annot=True, cmap='coolwarm', center=0)
plt.title('PCA Component Loadings')
plt.show()
```

```python
credit_score=[]
for index, row in df.iterrows():
    gend=row['Gender']
    loantype=row['Loan Type']
    dps=row['Defaulting Payment Status']
    debt=row['Debt']
    loanamount=row['Loan Amount']
    loanterm=row['Loan Term']
    credit_score.append((gend*0.30)+(loantype*0.30)+(dps*0.20)+(debt*0.1)+(loanterm*0.05)+(loanamount*0.05))
    
df['credit_score']=credit_score
df.head()
```

```python
df.credit_score = mmax.fit_transform(df[['credit_score']])*10
df.head()
```

```python
from sklearn.cluster import KMeans

X = df[['credit_score']]
kmeans = KMeans(n_clusters=3, n_init=10, random_state=42)
kmeans.fit(X)
df['Segment'] = kmeans.labels_
```

```python
import plotly.express as px

# Convert the 'Segment' column to category data type
df['Segment'] = df['Segment'].astype('category')

# Visualize the segments using Plotly
fig = px.scatter(df, x=df.index, y='credit_score', color='Segment',
                 color_discrete_sequence=['green', 'blue', 'yellow', 'red'])
fig.update_layout(
    xaxis_title='Customer Index',
    yaxis_title='Credit Score',
    title='Customer Segmentation based on Credit Scores'
)
fig.show()
```
